```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data
X = np.array([1, 2, 3, 4, 5 ]).reshape(-1, 1)  # variabel independen
y = np.array([2, 3, 4, 5, 4])  # variabel dependen
```

```python
# Membuat objek model
model = LinearRegression()

# Melatih model
model.fit(X, y)
```

```python
# Memprediksi nilai y berdasarkan X
y_pred = model.predict(X)

# Visualisasi data dan garis regresi
plt.scatter(X, y, color='blue', label='Data Asli')
plt.plot(X, y_pred, color='red', label='Garis Regresi')
plt.xlabel('Variabel Independen')
plt.ylabel('Variabel Dependen')
plt.title('Regresi Linear Sederhana')
plt.legend()
plt.show()
```
